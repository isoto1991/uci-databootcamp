```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
purchase_data.head()
```

```python
players = purchase_data[['SN', 'Age', 'Gender']].copy()
players = players.drop_duplicates()
pd.DataFrame({"Total Players": [players.count()[0]]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data["Item ID"].unique())
avg_price = (purchase_data["Price"].mean())
num_of_purchases = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()

purchase_analysis = pd.DataFrame({
    "Number of Unique Items": unique_items, 
    "Average Price": avg_price, 
    "Number of Purchases": num_of_purchases, 
    "Total Revenue": [total_revenue]
})

purchase_analysis.round(2)
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_demographics = players["Gender"].value_counts()
total_players = players.count()[0]
percentages = gender_demographics / total_players
gender_df = pd.DataFrame({
    "Total Count": gender_demographics, 
    "Percentage of Players": percentages
})
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count = purchase_data.groupby('Gender').Price.count()
average_purchase_price = purchase_data.groupby('Gender').Price.mean()
total_purchase_value = purchase_data.groupby('Gender').Price.sum()
total_purchase_per_person = total_purchase_value / purchase_count

purchasing_analysis_df = pd.DataFrame({
    "Purchase Count": purchase_count, 
    "Average Purchase Price": average_purchase_price, 
    "Total Purchase Value": total_purchase_value, 
    "Avg Total Purchase per person": total_purchase_per_person
})

purchasing_analysis_df
```

```python
bins = [0, 9.90, 19.9, 29.9, 39.9, 999]
bin_names = ["less than 10", "10-19", "20-29", "30-39", "40+"]

players['Player Age Demographics'] = pd.cut(players.Age, bins, labels=bin_names, right=True)

age_totals = players['Player Age Demographics'].value_counts()
age_percentages = age_totals / players.count()[0]

demographics_df = pd.DataFrame({
    "Total Count": age_totals, 
    "Percentage of Players": age_percentages
})

new_index = ["less than 10", "10-19", "20-29", "30-39", "40+"]
demographics_df = demographics_df.reindex(new_index)
demographics_df
```

# Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9.90, 19.9, 29.9, 39.9, 999]
bin_names = ["less than 10", "10-19", "20-29", "30-39", "40+"]

players['Player Age Demographics'] = pd.cut(players.Age, bins, labels=bin_names, right=True)

age_totals = players['Player Age Demographics'].value_counts()
age_percentages = age_totals / players.count()[0]

demographics_df = pd.DataFrame({
    "Total Count": age_totals, 
    "Percentage of Players": age_percentages
})

new_index = ["less than 10", "10-19", "20-29", "30-39", "40+"]
demographics_df = demographics_df.reindex(new_index)
demographics_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Purchasing Analysis(Age)"] = pd.cut(purchase_data["Age"], bins, labels=bin_names)

purchase_counts = purchase_data.groupby(["Purchasing Analysis(Age)"]).count()["Price"]
avg_purchase_price = purchase_data.groupby(["Purchasing Analysis(Age)"]).mean()["Price"]
total_purchase_value = purchase_data.groupby(["Purchasing Analysis(Age)"]).sum()["Price"]
avg_per_person = total_purchase_value / purchase_counts

age_analysis_df = pd.DataFrame({
    "Purchase Counts": purchase_counts, 
    "Average Purchase Price": avg_purchase_price, 
    "Total Purchase Value": total_purchase_value, 
    "Average Total Purchase Per Person": avg_per_person
})

age_analysis_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
user_purchase_count = purchase_data.groupby(["SN"]).count()["Price"]
user_avg_purchase = purchase_data.groupby(["SN"]).mean()["Price"]
user_total = purchase_data.groupby(["SN"]).sum()["Price"]

top_spender_df = pd.DataFrame({
    "Purchase Count": user_purchase_count, 
    "Average Purchase Price": user_avg_purchase, 
    "Total Purchase Value": user_total
})

top_spender_df = top_spender_df.sort_values(by=['Purchase Count'], ascending=False)

top_spender_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items = purchase_data[['Item ID', 'Item Name', 'Price']].copy()

item_count = popular_items.groupby(["Item ID", "Item Name"]).count()["Price"]
item_price = popular_items.groupby(["Item ID", "Item Name"]).mean()["Price"]
item_total = popular_items.groupby(["Item ID", "Item Name"]).sum()["Price"]

item_df = pd.DataFrame({
    "Purchase Count": item_count, 
    "Item Price": item_price, 
    "Total Purchase Value": item_total
})

item_df = item_df.sort_values(by=["Purchase Count"], ascending=False)
item_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
item_df_sorted = item_df.sort_values(by=['Total Purchase Value'], ascending=False)
item_df_sorted.head()
```
